```python
import pandas as pd
import os

# Load products data
products_file_name = 'products.csv'
products_file_path = os.path.join('/content/extracted_data', products_file_name)
df_products = pd.read_csv(products_file_path)

print(f"Successfully loaded {products_file_name}. Displaying the first 5 rows:")
display(df_products.head())
```

```python
# Merge df_sales with df_products on 'product_id'
df_sales = pd.merge(df_sales, df_products, on='product_id', how='left')

print("Sales data merged with product information. Displaying the first 5 rows of the merged DataFrame:")
display(df_sales.head())

print("Information about the merged df_sales DataFrame:")
df_sales.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate daily sales
daily_sales = df_sales.groupby('order_date')['revenue'].sum().reset_index()

plt.figure(figsize=(15, 7))
sns.lineplot(x='order_date', y='revenue', data=daily_sales)
plt.title('Daily Sales Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```
